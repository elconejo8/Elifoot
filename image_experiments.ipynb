```python
import pytesseract
from PIL import Image
import cv2
import numpy as np

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
def preprocess_dosbox_screen(image_path):
    """Preprocess DOSBox screenshot for better OCR"""
    
    # Read image
    img = cv2.imread(image_path)
    Image.fromarray(img).show()
    
    # Convert to grayscale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    Image.fromarray(img_gray).show()
    
    # Increase contrast and apply threshold
    # This will make text white on black background
    _, thresh = cv2.threshold(img_gray, 100, 255, cv2.THRESH_BINARY)
    
    # Alternative: Adaptive threshold (often better for DOS games)
    # thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    #                                cv2.THRESH_BINARY, 11, 2)
    
    # Invert if needed (text should be black on white for Tesseract)
    thresh = cv2.bitwise_not(thresh)
    Image.fromarray(thresh).show()

    # Optional: Denoise
    denoised = cv2.fastNlMeansDenoising(thresh)
    Image.fromarray(denoised).show()


    # define range of blue color in HSV
    lower_blue = np.array([10,50,50])
    upper_blue = np.array([13,255,255])
 
    # Threshold the HSV image to get only blue colors
    mask = cv2.inRange(img, lower_blue, upper_blue)
    Image.fromarray(mask).show()
    
    return img

# Usage
processed_image = preprocess_dosbox_screen(r'eli_image_example.png')

text = pytesseract.image_to_string(processed_image)
print(text)
```
