```python
import subprocess
import time
import pyautogui
from PIL import ImageGrab
import numpy as np
from dotenv import load_dotenv
import os
import pygetwindow as gw
```

```python
load_dotenv()

game_path = os.getenv('GAME_PATH')
dosbox_folder_path = os.getenv('DOSBOX_FOLDER')
```

```python
class EliBot:
    def __init__(self, game_exe, dosbox_path):
        """
        Launch game in DOSBox (DOS emulator)
        """
        # Start DOSBox with the game
        self.process = subprocess.Popen(
            game_path,
            cwd=dosbox_folder_path,
            stdin=subprocess.PIPE
        )
        time.sleep(3)  # Wait for game to load
        
        # Get game window position
        self.find_game_window()
    
    def find_game_window(self):
        """Find DOSBox window coordinates"""
        
        windows = gw.getWindowsWithTitle('DOSBox')
        if windows:
            self.game_window = windows[0]
            self.game_window.activate()
        else:
            print("Warning: Could not find game window")

    def send_key(self, key):
        """Send keyboard input to game"""
        print(f">>> Bot pressing: {key}")
        pyautogui.press(key)

    def play(self):
        self.send_key('n')
        time.sleep(25)
        self.send_key('G')
        time.sleep(25)
        self.send_key('enter')
```

```python
eb = EliBot(game_path, dosbox_folder_path)
eb.play()
```
